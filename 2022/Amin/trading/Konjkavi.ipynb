```python
import pandas as pd


df = pd.read_csv("EURUSD15.csv")
df[:10]
len(df)
```

```python
# macd
k = df['close'].ewm(span=12, adjust=False, min_periods=12).mean()  # blue line
d = df['close'].ewm(span=26, adjust=False, min_periods=26).mean()  # yellow line
macd = k - d

macd_s = macd.ewm(span=9, adjust=False, min_periods=9).mean()

macd_h = macd - macd_s

# macd 1
k = df['close'].ewm(span=48, adjust=False, min_periods=48).mean()  # blue line
d = df['close'].ewm(span=104, adjust=False, min_periods=104).mean()  # yellow line

macd1 = k - d
macd_s1 = macd1.ewm(span=36, adjust=False, min_periods=36).mean()
macd_h1 = macd1 - macd_s1

df['macd'] = df.index.map(macd)
df['macd_h'] = df.index.map(macd_h)
df['macd_s'] = df.index.map(macd_s)

df['macd1'] = df.index.map(macd1)
df['macd_h1'] = df.index.map(macd_h1)
df['macd_s1'] = df.index.map(macd_s1)
```

```python
diff_macd = []

for index, row in df.iterrows():
    macd = row["macd"]
    macd1 = row["macd1"]
    diff = (macd < 0 < macd1) or (macd1 < 0 < macd)
    
    if diff:
        diff_macd.append(row)
```
