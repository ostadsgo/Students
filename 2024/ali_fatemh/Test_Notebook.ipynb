```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('nba.csv')

# Extract Age and Salary columns
age = df['Age']
salary = df['Salary']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, salary, color='blue', alpha=0.5)  # alpha parameter controls transparency
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('nba.csv')

# Extract Age and Salary columns
age = df['Age']
salary = df['Salary']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, salary, color='blue', alpha=0.5)  # alpha parameter controls transparency
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
df.columns
```

```python
# 1. Salary Distribution (Histogram)
plt.figure(figsize=(10, 6))
plt.hist(df['Salary'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution of NBA Players')
plt.grid(True)
plt.show()
```

```python
# 2. Position Distribution (Bar chart)
position_counts = df['Position'].value_counts()
plt.figure(figsize=(10, 6))
position_counts.plot(kind='bar', color='salmon')
plt.xlabel('Position')
plt.ylabel('Number of Players')
plt.title('Distribution of NBA Players by Position')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
# 3. Age vs. Salary Relationship (Scatter plot)
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], color='orange', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs. Salary of NBA Players')
plt.grid(True)
plt.show()
```
